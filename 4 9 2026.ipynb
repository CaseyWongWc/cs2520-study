# Summary of Lecture Notes on Python Class and Type Checking

## Key Concepts
- **`isinstance()` function**: A built-in Python function used to check if an object is an instance of a specified class, which can be applied to both built-in types and custom objects.
- **`issubclass()` function**: A built-in function that checks if a class is a direct or indirect subclass of another class. 

## `isinstance()` Function

### Usage
- Syntax: `isinstance(obj, class)` checks if `obj` is an instance of `class`.
- Example with **Built-in Types**:
    ```python
    # Lists
    print(isinstance([1, 2, 3], list))  # ✅ True
    print(isinstance((1, 2, 3), list))  # ❌ False (tuple)

    # Strings
    print(isinstance("123", str))        # ✅ True
    print(isinstance(123, str))          # ❌ False

    # Checking Multiple Types
    value = 42
    print(isinstance(value, (int, str)))  # ✅ True
    print(isinstance(value, (str, list)))  # ❌ False
    ```

### Usage with Custom Objects
- Example with **Class Inheritance**:
    ```python
    class Parent:
        pass

    class Child(Parent):
        pass

    p = Parent()
    c = Child()

    print(isinstance(c, Child))  # ✅ True
    print(isinstance(c, Parent))  # ✅ True
    print(isinstance(p, Child))   # ❌ False
    print(isinstance(c, Other))   # ❌ False
    ```

- *Inheritance Explanation*:
    - An instance of a subclass is also considered an instance of its superclass.
    - Conversely, a superclass instance is not an instance of its subclass.

## `issubclass()` Function

### Usage
- Syntax: `issubclass(subclass, parent_class)` checks if `subclass` is a child class of `parent_class`.
- Example:
    ```python
    class Parent:
        pass

    class Child(Parent):
        pass

    class GrandChild(Child):
        pass

    class Other:
        pass

    print(issubclass(Child, Parent))       # ✅ True
    print(issubclass(GrandChild, Child))   # ✅ True
    print(issubclass(GrandChild, Parent))   # ✅ True
    print(issubclass(Parent, Child))       # ❌ False
    print(issubclass(Other, Parent))       # ❌ False
    ```

## Conclusion
- Understanding **type checking** with `isinstance()` and **class hierarchy** with `issubclass()` is fundamental for working effectively with Python classes and objects, especially when interfacing with external libraries or implementing methods in inherited classes.

````python
#Invoke Parent's Class Methods in Child Class

#The child class provides its own version of say_hi() that extends the parent's behavior.

class Parent:
    def say_hi(self):
        return "Hola!"

class Child(Parent):
    def say_hi(self): # override the inherited methods
        # return Parent.say_hi(self) + " Hi!"
        return super().say_hi() + " Hi!"

c = Child()
print(c.say_hi()) # "Hola! Hi!"
p = Parent()
print(p.say_hi()) # "Hola!"
````

````python
class Parent:
    def say_hi(self):
        return "Hola!"

class Child(Parent):
    def say_hi(self):
        return Parent.say_hi(self) + " Hi!"

print(Child().say_hi())
````

````python
class Parent:
    def __init__(self, name, languages):
        self.name = name
        self.languages = languages
    def __str__(self):
        return f'{self.name}, {self.languages}'

class Child(Parent):
    def __init__(self, name, languages, religion):
        Parent.__init__(self, name, languages)
        self.religion = religion
    def __str__(self):
        return super().__str__() + f'.Religion: {self.religion}'
````
