```python
# ── OS & PATH HELPERS ──────────────────────────
import os

# See where you currently are
print("Current directory:", os.getcwd())

# List files in current folder
print("\nFiles here:")
for f in os.listdir("."):
    print(" ", f)

# Make a new folder (won't error if it already exists)
os.makedirs("test_output", exist_ok=True)
print("\n✅ 'test_output/' folder ready")
```

```python
# ── WRITE A FILE ───────────────────────────────

# Method 1: Python open() — classic
with open("test_output/hello.txt", "w") as f:
    f.write("Hello from Python!\n")
    f.write("This is line 2.\n")

print("✅ hello.txt written")
```

```python
# ── READ A FILE ────────────────────────────────

# Read entire file at once
with open("test_output/hello.txt", "r") as f:
    contents = f.read()

print("Full file:\n", contents)

# Read line by line
with open("test_output/hello.txt", "r") as f:
    for i, line in enumerate(f):
        print(f"  Line {i}: {line.strip()}")
```

```python
# ── APPEND & FILE EXISTENCE ────────────────────

# Check if a file exists before doing stuff
path = "test_output/hello.txt"
if os.path.exists(path):
    print("✅ File found:", path)
else:
    print("❌ File not found")

# Append to a file (won't overwrite!)
with open(path, "a") as f:
    f.write("Appended line! 🎉\n")

# Verify the append worked
with open(path, "r") as f:
    print("\nUpdated file:")
    print(f.read())
```
