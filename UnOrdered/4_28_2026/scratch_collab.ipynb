#### How to use `parse`

```python
_ = Problem("my problem")
_.parse('''
F:
def greet():
    return "hi"
F:: greet.py

F:
import greet
print(greet.greet())
F:: main.py

R: main.py
''')
```

- `F:` starts a file block
- `F:: filename.py` ends the block and saves the file
- `FF: filename.py` is also a valid end marker
- `F:: filename.py` can appear first, and `F:` can close it (reversed style)
- You can also put content first, then end with `F:: filename.py` (start marker optional)
- `R: main.py` runs the file
- `R: main.py ["Eve", "5"]` runs with those inputs

````python
# Demo — challenge activity 11.3.1 rewritten with parse
_ = Problem("challenge 11.3.1 demo")
_.parse('''
F:
def order_potato_pie(buyer_name, potato_quantity):
    print(f"{buyer_name} orders a pie with {potato_quantity} potatoes.")
F:: pie_shop.py

F:
def bake(potato_quantity):
    print(f"{potato_quantity} potatoes become one potato pie.")

def decorate(potato_quantity):
    print(f"{potato_quantity} potatoes are used to decorate one potato pie.")
F:: potato_pie.py

F:
from pie_shop import order_potato_pie
from potato_pie import bake, decorate

buyer_name = input()
potato_quantity = int(input())

order_potato_pie(buyer_name, potato_quantity)
bake(potato_quantity - 2)
decorate(4)
F:: main.py

R: main.py ["Eve", "5"]
''')
````
