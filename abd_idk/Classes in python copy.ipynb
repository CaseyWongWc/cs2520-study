````python
'''Python Attributes

In Python, attributes are properties associated with an object or a class, accessible using dot notation (e.g., object.attribute). They store data and define the state of an object.

There are 2 types of attributes:

[1] Class Attributes:
Class attributes are variables that are shared among all instances of a class. They belong to the class itself, not a specific instance, and are useful for storing data that should be the same for every object, such as a default value or a constant.

[2] Instance Attributes
Instance attributes are unique to each individual object (instance) created from a class. They define the specific characteristics of an object and can have different values for different instances of the same class.'''
````

````python
'''Python Class Attributes  

1. They are defined within the class body, but outside of any instance methods (like `__init__`).  

2. Scope: They are shared across all instances, using a single memory location.  

3. Access: They can be accessed using either the class name or any instance of the class (e.g., Car.wheels or car1.wheels).  

4. Modification: Modifying a class attribute through the class name affects all instances, while modifying it through an instance creates a new instance attribute that "shadows" the class attribute for that specific instance.'''
````

````python
'''Modify Mutable Class Attributes

class Bag:
    items = []  # class attribute

b1 = Bag()
b2 = Bag()

b1.items.append("apple")

print(b1.items)  # ['apple']
print(b2.items)  # ['apple']
print(Bag.items)  # ['apple']

List is a mutable object. Thus,
The items list is created once, and all instances of the Bag class will point to the same list in memory

b1.items
    └──> SAME LIST items []
b2.items

How to fix it? turn the mutable object to be an instance attribute, instead'''
````

````python
def slide10():
    class Bag:
        items = []  # class attribute

    b1 = Bag()
    b2 = Bag()

    b1.items.append("apple")

    print(b1.items)  # ['apple']
    print(b2.items)  # ['apple']
    print(Bag.items)  # ['apple']
slide10()
"-"
def slide10_1():
    class Bag:
        def __init__(self):
            self.items = []  # instance attribute

    b1 = Bag()
    b2 = Bag()

    b1.items.append("apple")

    print(b1.items)  # ['apple']
    print(b2.items)  # []
    #print(Bag.items)  # AttributeError: type object 'Bag' has no attribute 'items'
slide10_1()
````

````python
'''**Instance Methods**  

An instance method is a function defined inside a class that works with a specific object (instance) of that class.  

- **Self parameter:** instance methods always take self as its first parameter. When the method is called on an instance (e.g., my_bject.my_method()), Python automatically passes that instance object as the first argument to the self parameter.
- **Access to instance data:** The self parameter allows the method to access and manipulate the instance's attributes, which are unique to that particular object.
- **Access to class data:** Instance methods can also access class attributes using self.__myClass__.__class_attribute or by directly referencing the class name such as myClass.class_attribute, allowing them to potentially modify the class state as well.'''
````

````python
'''Instance Method Example
class BankAccount:
    def __init__(self, balance):
        self.balance = balance

    def deposit(self, amount):
        if amount <= 0:
            return "Deposit must be positive"
        self.balance += amount
        return f"New balance: {self.balance}"

    def withdraw(self, amount):
        if amount <= 0:
            return "Withdrawal must be positive"
        if amount > self.balance:
            return "Invalid transaction"
        
        self.balance -= amount
        return f"New balance: {self.balance}"

acc = BankAccount(100)
print(acc.deposit(50))
print(acc.withdraw(30))'''
````

````python
'''Converting Objects to a String

class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def __str__(self):
        return f"Width: {self.width} - Height: {self.height}"

r1 = Rectangle(4, 3)
print(r1) # Width: 4 - Height: 3

Notice that __str__ method uses the same naming convention as the constructor, that is two underscores before and after the name. It is common that Python uses this naming technique for special methods.

The __str__ method returns a string representation as defined by the class creator.'''
````

````python
'''Object as An Argument

We can pass an object as an argument to a function

class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

def areaDifference(rect1, rect2):
    diff = abs(rect1.area() - rect2.area())
    return diff

r1 = Rectangle(4, 3)
r2 = Rectangle(2, 1)
diffArea = areaDifference(r1, r2)
print(diff)  # 10

Note that areaDifference is not a method of the Rectangle class.
It is not inside the class definition.

The other way we can know that areaDifference is not a method of Rectangle is that self is not included in the 1st parameter.

Thus, we do not invoke areaDifference using the dot notation.

Can we move areaDifference inside the class as an instance method, instead of keeping it as a standalone function outside?'''
````

````python
'''Object as a Parameter

class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

    def areaDifference(self, anotherRec):
        diff = abs(self.area() - anotherRec.area())
        return diff

r1 = Rectangle(4, 3)
r2 = Rectangle(2, 1)
diffArea = r1.areaDifference(r2)
print(diffArea) # 10'''
````
