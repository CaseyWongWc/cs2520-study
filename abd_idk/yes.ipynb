````python
import builtins

# Store the original input function once, globally, to avoid re-capturing a mocked input
if not hasattr(builtins, '_original_input_backup'):
    builtins._original_input_backup = builtins.input

def set_test_inputs(inputs_list):
    """
    Sets up a mock input function that draws from a list of inputs.
    If inputs_list is None or empty, restores the original input function.
    """
    if inputs_list:
        input_iter = iter(inputs_list)
        def mock_input(prompt=""):
            try:
                value = next(input_iter)
                print(f"{prompt}{value}")
                return value
            except StopIteration:
                raise EOFError("No more inputs for testing")
        builtins.input = mock_input
    else:
        builtins.input = builtins._original_input_backup

# Define test_inputs, this is where you can change your desired inputs
test_inputs = [1]  # <--- Change this list to provide your test inputs

# Apply the test inputs automatically when this cell is run
set_test_inputs(test_inputs)

# Example of how to revert to normal input:
# set_test_inputs(None)
# Or to set new inputs dynamically:
# set_test_inputs(["new_input_1", "new_input_2"])

#===

def setin(*inputs):
    """
    A helper function to set test inputs for the input() function.
    Usage:
    setin("input1", "input2", "input3")
    This will set up the input() function to return "input1" on the first call,
    "input2" on the second call, and so on.
    To reset to normal input behavior, call setin() with no arguments or None:
    setin()
    """
    if inputs:
        input_iter = iter(inputs)
        def mock_input(prompt=""):
            try:
                value = next(input_iter)
                print(f"{prompt}{value}")
                return value
            except StopIteration:
                raise EOFError("No more inputs for testing")
        builtins.input = mock_input
    else:
        builtins.input = builtins._original_input_backup
````

````python
'''Python Class

Everything we’ve learned so far is procedure programming. But, Python 
is an object-oriented programming language like Java, C++. That means it 
provides features that support object-oriented programming (OOP).

A class is a blueprint/template to create objects (instances). Each class 
has attributes (or properties) and methods (functionalities)

In Python, everything is an object either string, float, integer, or list. 
These are pre-defined objects. But, we can also create our own 
classes in Python.'''
````

````python
'''Python Class

Instantiating an object means creating an instance of a class.

Class = blueprint
Object (instance) = actual thing created from that blueprint

class Employee:
    pass

my_employee = Employee() # instantiating

What happens during instantiation?
- Creates a new empty object
- Calls the class constructor (__init__, if it exists)
- Returns the object and assigns it to my_employee'''
````

````python
'''Python Attributes

In Python, attributes are properties associated with an object or a class, accessible using dot notation (e.g., object.attribute). They store data and define the state of an object.

There are 2 types of attributes:

[1] Class Attributes:
Class attributes are variables that are shared among all instances of a class. They belong to the class itself, not a specific instance, and are useful for storing data that should be the same for every object, such as a default value or a constant.

[2] Instance Attributes
Instance attributes are unique to each individual object (instance) created from a class. They define the specific characteristics of an object and can have different values for different instances of the same class.'''
````

````python
'''Python Class Attributes  

1. They are defined within the class body, but outside of any instance methods (like `__init__`).  

2. Scope: They are shared across all instances, using a single memory location.  

3. Access: They can be accessed using either the class name or any instance of the class (e.g., Car.wheels or car1.wheels).  

4. Modification: Modifying a class attribute through the class name affects all instances, while modifying it through an instance creates a new instance attribute that "shadows" the class attribute for that specific instance.'''
````

````python
'''Python Class Attributes

class Dog:
    # class attribute
    age = 1

dog1 = Dog()
dog2 = Dog()

print(dog1.age)  # 100
print(dog2.age)  # 100
print(Dog.age)   # 100'''
````

````python
'''Modify Immutable Class Attributes

class Dog:
    # class attribute
    age = 1

dog1 = Dog()
dog2 = Dog()
Dog.age = 100  # modify class attribute
print(dog1.age) # 100
print(dog2.age) # 100
print(Dog.age)  # 100

NOTE: Changes to a class attribute will affect all instances
of the class, as they share the same class attribute'''
````

````python
'''Modify Immutable Class Attributes

class Dog:
    # class attribute
    age = 1

dog1 = Dog()
dog2 = Dog()

dog1.age = 10  # override value of the instance attribute
                # but NOT the class attribute

print(dog1.age)  # 10
print(dog2.age)  # STILL 1
print(Dog.age)   # STILL 1

NOTE: Changes to a class attribute will affect all instances of the class, as they share the same class attribute unless it is overridden by an instance attribute.'''
````

````python
'''Modify Mutable Class Attributes

class Bag:
    items = []  # class attribute

b1 = Bag()
b2 = Bag()

b1.items.append("apple")

print(b1.items)  # ['apple']
print(b2.items)  # ['apple']
print(Bag.items)  # ['apple']

List is a mutable object. Thus,
The items list is created once, and all instances of the Bag class will point to the same list in memory

b1.items
    └──> SAME LIST items []
b2.items

How to fix it? turn the mutable object to be an instance attribute, instead'''
````

````python
'''Modify Class Attributes

NOTE:  

Changes to a class attribute will affect all instances of the class, as they share the same class attribute unless it is overridden by an instance attribute. However, if the class attribute is a mutable object (e.g., list, dictionary, set), modifying it will affect all instances.'''
````

````python
'''Instance Attribute

1. Definition: They are typically defined within the class's initializer method (or constructor), __init__, using the self keyword (e.g., self.name = name).

2. Scope: Each instance has its own copy of the instance attribute in memory.

3. Access: They are accessed using the instance object (e.g., dog.name).

4. Modification: Changing an instance attribute only affects that specific instance.'''
````

````python
'''Instance Attribute Example
class Dog:
    def __init__(self, name):
        self.name = name # instance attribute

d1 = Dog("Buddy")
d2 = Dog("Max")

print(d1.name) # Buddy
print(d2.name) # Max

Class attribute = shared among all objects
Instance attribute = unique per object'''
````

````python
'''Constructor
```
class Dog:
    def __init__(self, name): # constructor
        self.name = name # instance attribute

d1 = Dog("Buddy")
```
__init__ in Python is a special method (also known as a constructor) that runs automatically when you create a new object from a class. It set up its initial state of an object.

self is a convention for self-reference by the instance object.

Explain the code
__init__ is automatically called by Python when Dog("Buddy") runs
- "Buddy" is passed into __init__
- self.name store the value in the object'''
````

````python
'''Instance Attribute Example
How to Avoid the Problem with modifying mutable class attribute?
Ans: Using instance attribute

class Bag:
    def __init__(self):
        self.items = [] # separate list per object

b1 = Bag()
b2 = Bag()

b1.items.append("apple")
print(b1.items) # ['apple']
print(b2.items) # []'''
````
