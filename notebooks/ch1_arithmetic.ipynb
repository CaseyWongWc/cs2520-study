## 🔢 The 4 Basic Operators
| Operator | Name | Example | Result |

## ⚡ Special Operators
| Operator | Name | Example | Result |
|----------|------|---------|--------|
| `**` | Exponentiation | `2 ** 4` | `16` |
| `//` | Integer (Floor) Division | `7 // 2` | `3` |
| `%` | Modulo (remainder) | `7 % 2` | `1` |

**The `%` operator** gives you the **leftover** after division. Super useful for:
- Checking if a number is even/odd (`n % 2 == 0` means even!)
- Wrapping numbers around (like clock math)

````python
print('2 ** 4  =', 2 ** 4)    # 16   (2 to the power of 4)
print('7 // 2  =', 7 // 2)   # 3    (drops the decimal)
print('7 % 2   =', 7 % 2)    # 1    (remainder: 7 = 3*2 + 1)
print('-4 ** 2 =', -4 ** 2)  # -16  (⚠️ tricky! see below)
````

````python
print(2 ** 4)
print(7//2)
print(7%2)
print(-4 ** 2)
````

## 📋 Operator Precedence (PEMDAS)
Python follows math order-of-operations. From **highest to lowest priority**:

| Priority | Operator | Name |
|----------|----------|------|
| 1 (highest) | `()` | Parentheses |
| 2 | `**` | Exponentiation |
| 3 | `-x` | Negative (unary) |
| 4 | `* / // %` | Multiply, Divide |
| 5 (lowest) | `+ -` | Add, Subtract |

> ⚠️ **Tricky**: `-4 ** 2` = `-(4**2)` = **-16**, NOT `(-4)**2` = 16!
> Use parentheses `(-4) ** 2` to be safe!

````python
# From your lecture slides!
print((1 + 2) * 3)   # → 9   (parens first)
print(1 + 2 * 3)     # → 7   (* before +)
print(2 ** 4)        # → 16
print(-4 ** 2)       # → -16  (tricky!)
print((-4) ** 2)     # → 16   (parentheses save us!)
````

## 🕐 The % Operator — Modulo Deep Dive
Think of it like clock math or division with a remainder.

```
13 % 4 = 1   because   13 = 3 × 4 + 1
10 % 3 = 1   because   10 = 3 × 3 + 1
10 % 2 = 0   because   10 = 5 × 2 + 0   ← even number!
```

````python
# Modulo examples
print(13 % 4)    # 1
print(10 % 3)    # 1
print(10 % 2)    # 0 (even!)
print(7 % 2)     # 1 (odd!)

# SUPER useful pattern: check even/odd
n = 42
if n % 2 == 0:
    print(n, 'is EVEN')
else:
    print(n, 'is ODD')
````
