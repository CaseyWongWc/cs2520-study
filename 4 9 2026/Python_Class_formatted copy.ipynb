## Python Attributes

In Python, attributes are properties associated with an object or a class, accessible using dot notation (e.g., object.attribute). They store data and define the state of an object.

There are 2 types of attributes:

## [1] Class Attributes:

Class attributes are variables that are shared among all instances of a class. They belong to the class itself, not a specific instance, and are useful for storing data that should be the same for every object, such as a default value or a constant.

## [2] Instance Attributes

Instance attributes are unique to each individual object (instance) created from a class. They define the specific characteristics of an object and can have different values for different instances of the same class



## Python Class Attributes

1. They are defined within the class body, but outside of any instance methods (like **init**).
2. Scope: They are shared across all instances, using a single memory location.
3. Access: They can be accessed using either the class name or any instance of the class (e.g., Car.wheels or car1.wheels).
4. Modification: Modifying a class attribute through the class name affects all instances, while modifying it through an instance creates a new instance attribute that “shadows” the class attribute for that specific instance.



## Python Class Attributes

````python
class Dog:
    # class attribute
    age = 1
dog1 = Dog()
dog2 = Dog()
print(dog1.age) # 100
print(dog2.age) # 100
print(Dog.age) # 100
````

## Modify Immutable Class Attributes

````python
class Dog:
    # class attribute
    age = 1
dog1 = Dog()
dog2 = Dog()
Dog.age = 100 # modify class attribute
print(dog1.age) # 100
print(dog2.age) # 100
print(Dog.age) # 100
````

NOTE: Changes to a class attribute will affect all instances of the class, as they share the same class attribute

## Modify Mutable Class Attributes

````python
class Bag:
    items = [] # class attribute
b1 = Bag()
b2 = Bag()
b1.items.append("apple")
print(b1.items) # ['apple']
print(b2.items) # ['apple']
print(Bag.items) # ['apple']
````

List is a mutable object. Thus, The items list is created once, and all instances of the Bag class will point to the same list in memory

````python
b1 = Bag()
b2 = Bag()
````


b1.items → SAME LIST items []

How to fix it? turn the mutable object to be an instance attribute, instead

## Instance Attribute



1. Definition: They are typically defined within the class’s initializer method (or constructor),`**init`**, using the self keyword (e.g., self.name = name).
2. Scope: Each instance has its own copy of the instance attribute in memory.
3. Access: They are accessed using the instance object (e.g., dog.name).
4. Modification: Changing an instance attribute only affects that specific instance.



## Instance Attribute Example

````python
class Dog:
    def __init__(self, name):
        self.name = name # instance attribute
d1 = Dog("Buddy")
d2 = Dog("Max")
print(d1.name) # Buddy
print(d2.name) # Max
#Class attribute = shared among all objects
#Instance attribute = unique per object
````

## Instance Methods




An instance method is a function defined inside a class that works with a specific object (instance) of that class.

- Self parameter: instance methods always take self as its first parameter. When the method is called on an instance (e.g., my_bject.my_method()), Python automatically passes that instance object as the first argument to the self parameter.
- Access to instance data: The self parameter allows the method to access and manipulate the instance’s attributes, which are unique to that particular object.
- Access to class data: Instance methods can also access class attributes using self.__myClass__.class_attribute or by directly referencing the class name such as myClass.class_attribute, allowing them to potentially modify the class state as well.



## Instance Method Example 1

````python
class Dog:
    def __init__(self, name):
        self.name = name
    def bark(self): # instance method
        print(f"{self.name} says woof!")
d1 = Dog("Buddy")
d1.bark()
````

## Instance Method Example 2

````python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height
    def area(self):
        return self.width * self.height
    def describe(self):
        print(f"Rectangle {self.width}x{self.height}, area {self.area()}")
r = Rectangle(4, 5)
r.describe()
````

## Instance Method Example

````python
class BankAccount:
    def __init__(self, balance):
        self.balance = balance
    def deposit(self, amount):
        if amount <= 0:
            return "Deposit must be positive"
        self.balance += amount
        return f"New balance: {self.balance}"
    def withdraw(self, amount):
        if amount <= 0:
            return "Withdrawal must be positive"
        if amount > self.balance:
            return "Invalid transaction"
        self.balance -= amount
        return f"New balance: {self.balance}"
acc = BankAccount(100)
print(acc.deposit(50)) # New balance: 150
print(acc.withdraw(30)) # New balance: 120
print(acc.withdraw(200)) # Invalid transaction
print(acc.deposit(-20)) # Deposit must be positive
print(acc.withdraw(-10)) # Withdrawal must be positive
````

## Converting Objects to a String

````python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height
    def __str__(self):
        return f"Width: {self.width} - Height: {self.height}"
r1 = Rectangle(4, 3)
print(r1) # Width: 4 - Height: 3
````





Notice that ``__str__`` method uses the same naming convention as the constructor, that is two underscores before and after the name. It is common that Python uses this naming technique for special methods.

The ``__str__`` method returns a string representation as defined by the class creator.



## Object as An Argument

We can pass an object as an argument to a function

````python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height
    def area(self):
        return self.width * self.height
def areaDifference(rect1, rect2):
    diff = abs(rect1.area() - rect2.area())
    return diff
r1 = Rectangle(4, 3)
r2 = Rectangle(2, 1)
diffArea = areaDifference(r1, r2)
print(diffArea) # 10  (it was diff before but I changed it to diffArea to avoid confusion with the variable name diff in the areaDifference function)
````

Note that areaDifference is not a method of the Rectangle class.

It is not inside the class definition.
The other way we can know that areaDifference is not a method of Rectangle is that self is not included in the 1st parameter.

Thus, we do not invoke areaDifference using the dot notation.

Can we moved areaDifference inside the class as an instance method, instead of keeping it as a standalone function outside?



## Object as a Parameter

````python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height
    def area(self):
        return self.width * self.height
    def areaDifference(self, anotherRec):
        diff = abs(self.area() - anotherRec.area())
        return diff
r1 = Rectangle(4, 3)
r2 = Rectangle(2, 1)
diffArea = r1.areaDifference(r2)
print(diffArea) # 10
````

Swap two numbers

````python
class Color:
    def __init__(self, initR, initG, initB):
        self.r = initR
        self.g = initG
        self.b = initB
    def __str__(self):
        return f"r = {self.r}, g = {self.g}, b = {self.b}"
    def blend(self, otherColor):
        mr = (self.r + otherColor.r) // 2
        mg = (self.g + otherColor.g) // 2
        mb = (self.b + otherColor.b) // 2
        # creates and returns a NEW Color object
        return Color(mr, mg, mb)
c1 = Color(100, 200, 50)
c2 = Color(200, 100, 150)
mixed = c1.blend(c2)
print(mixed) # r = 150, g = 150, b = 100
print(mixed.r) # 150
print(mixed.g) # 150
print(mixed.b) # 100
````
