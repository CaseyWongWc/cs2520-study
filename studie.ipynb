````python
qwertyuiop="""

# CS 2520 — Python for Programmers: Midterm Master Study Guide
## Chapters 1–7 | Prof. Nhi Nguyen | Spring 2026

---

# PART 1: COMPREHENSIVE TOPIC REVIEW

---

## Chapter 1 — Python Basics & Getting Started

### Key Concepts
- Python is **interpreted** (line-by-line), **dynamically typed** (variables can change type)
- Python is **slower** than Java because: dynamic typing, GIL (Global Interpreter Lock), no JIT compiler by default
- Python programs are **shorter** than Java (simpler syntax, less boilerplate)
- Source files use `.py` extension — called **script files** or **modules**
- Run scripts: `python demo.py` (run and exit) or `python -i demo.py` (run then interactive mode)
- `python -c "code"` runs one-liner from terminal without creating a file

### print() Function
```python
print("Hello")
print("Hello", "World")       # prints: Hello World (space separated)
print("Hello", end="")        # no newline at end
print("A", "B", sep="-")      # prints: A-B (custom separator)
print()                        # prints blank line
```

### input() Function
```python
name = input("Enter name: ")  # always returns a STRING
age = int(input("Age: "))     # must cast to int for numbers
price = float(input("Price: "))  # cast to float for decimals
```

### Comments
```python
# Single line comment
x = 5  # Inline comment

# Multi-line comments use multiple # signs
# There is no built-in multi-line comment syntax
# (triple quotes are docstrings, not true comments)
```

### Errors
| Error Type | When | Example |
|-----------|------|---------|
| **Syntax Error** | Code breaks grammar rules | `print("hello"` (missing `)`) |
| **Runtime Error** | Crashes during execution | `5 / 0` (ZeroDivisionError) |
| **Logic Error** | Wrong result, no crash | Using `+` instead of `*` |

### Escape Characters
| Escape | Meaning | Example |
|--------|---------|---------|
| `\n` | Newline | `print("Hi\nBye")` → two lines |
| `\t` | Tab | `print("Col1\tCol2")` |
| `\\` | Literal backslash | `print("C:\\Users")` |
| `\'` | Single quote in single-quoted string | `print('it\'s')` |
| `\"` | Double quote in double-quoted string | `print("He said \"hi\"")` |

---

## Chapter 2 — Variables, Types & Expressions

### Variables & Assignment
```python
x = 5          # assignment
x = x + 1      # update (x is now 6)
x += 1         # shorthand (x is now 7)
x, y = 10, 20  # multiple assignment
x, y = y, x    # swap values!
```

### Data Types
| Type | Example | Notes |
|------|---------|-------|
| `int` | `42`, `-7`, `0` | Unlimited precision in Python |
| `float` | `3.14`, `-0.5`, `2.0` | Floating-point (may have precision issues) |
| `str` | `"hello"`, `'world'` | Immutable sequence of characters |
| `bool` | `True`, `False` | Subclass of int (True=1, False=0) |
| `None` | `None` | Represents "no value" |

### Type Conversion
```python
int("42")       # str → int: 42
float("3.14")   # str → float: 3.14
str(42)         # int → str: "42"
int(3.9)        # float → int: 3 (TRUNCATES, does NOT round!)
int("3.5")      # ERROR! Can't convert float string directly to int
float("42")     # str → float: 42.0
bool(0)         # False  (0, 0.0, "", [], None are all False)
bool(1)         # True   (any non-zero/non-empty is True)
```

### Arithmetic Operators (in precedence order)
| Operator | Description | Example | Result |
|----------|-------------|---------|--------|
| `()` | Parentheses | `(1 + 2) * 3` | `9` |
| `**` | Exponentiation | `2 ** 4` | `16` |
| `+x`, `-x` | Unary positive/negative | `-math.pi` | `-3.14159` |
| `*`, `/`, `//`, `%` | Multiply, divide, floor divide, modulo | `7 // 2` | `3` |
| `+`, `-` | Addition, subtraction | `1 + 2` | `3` |

### Division Types
| Operation | Operator | Example | Result | Type |
|-----------|----------|---------|--------|------|
| True division | `/` | `7 / 2` | `3.5` | float |
| Floor division | `//` | `7 // 2` | `3` | int |
| Modulo (remainder) | `%` | `7 % 2` | `1` | int |
| Exponentiation | `**` | `2 ** 3` | `8` | int |

**Key gotcha**: `-4 ** 2` = `-16` (not 16!) because `**` has higher precedence than unary `-`
Use `(-4) ** 2` = `16` if you want the square of -4.

### Compound Assignment Operators
| Operator | Equivalent | Example |
|----------|-----------|---------|
| `+=` | `x = x + y` | `x += 3` |
| `-=` | `x = x - y` | `x -= 3` |
| `*=` | `x = x * y` | `x *= 3` |
| `/=` | `x = x / y` | `x /= 3` |
| `//=` | `x = x // y` | `x //= 3` |
| `%=` | `x = x % y` | `x %= 3` |
| `**=` | `x = x ** y` | `x **= 3` |

### Useful Modules
```python
import math
math.sqrt(16)    # 4.0
math.ceil(3.2)   # 4
math.floor(3.9)  # 3
math.pi          # 3.14159...
math.e           # 2.71828...
math.pow(2, 3)   # 8.0 (always returns float)
math.fabs(-5)    # 5.0 (absolute value as float)

import random
random.randint(1, 10)        # random int from 1 to 10 (inclusive)
random.random()              # random float 0.0 to 1.0
random.choice([1, 2, 3])     # random element from list
random.seed(42)              # set seed for reproducibility
```

### Naming Rules (Identifiers)
- Must start with a letter or underscore `_`
- Can contain letters, digits, underscores
- **Case-sensitive**: `myVar` ≠ `myvar`
- Cannot use Python keywords (`if`, `for`, `class`, `def`, etc.)
- Convention: `snake_case` for variables/functions, `UPPER_CASE` for constants

---

## Chapter 3 — Strings, Lists, Tuples, Sets, Dictionaries

### Strings (Immutable)
```python
s = "Hello World"
len(s)          # 11
s[0]            # 'H' (first character)
s[-1]           # 'd' (last character)
s[0:5]          # 'Hello' (slice: start to end-1)
s[:5]           # 'Hello' (omit start = from beginning)
s[6:]           # 'World' (omit end = to end)
s[::2]          # 'HloWrd' (every 2nd character)
s[::-1]         # 'dlroW olleH' (reverse!)
```

### String Formatting (f-strings & format specifiers)
```python
name = "Casey"
age = 21
gpa = 3.856

# f-string (preferred method)
print(f"Name: {name}, Age: {age}")
print(f"GPA: {gpa:.2f}")       # 3.86 (2 decimal places)
print(f"{'hello':>10}")         # "     hello" (right-align, width 10)
print(f"{'hello':<10}")         # "hello     " (left-align, width 10)
print(f"{'hello':^10}")         # "  hello   " (center-align, width 10)
print(f"{42:05d}")              # "00042" (zero-padded, width 5)
print(f"{0.5:.0%}")             # "50%" (percentage)
```

### Format Specifiers Table
| Specifier | Meaning | Example | Result |
|-----------|---------|---------|--------|
| `d` | Integer | `f"{42:d}"` | `42` |
| `f` | Float (fixed) | `f"{3.14159:.2f}"` | `3.14` |
| `e` | Scientific notation | `f"{1234:.2e}"` | `1.23e+03` |
| `s` | String | `f"{'hi':>10s}"` | `"        hi"` |
| `%` | Percentage | `f"{0.85:.1%}"` | `85.0%` |
| `<` | Left-align | `f"{'hi':<10}"` | `"hi        "` |
| `>` | Right-align | `f"{'hi':>10}"` | `"        hi"` |
| `^` | Center-align | `f"{'hi':^10}"` | `"    hi    "` |
| `0` | Zero-pad | `f"{7:03d}"` | `007` |

### Lists (Mutable, Ordered)
```python
nums = [1, 2, 3, 4, 5]
nums[0]          # 1
nums[-1]         # 5
nums[1:3]        # [2, 3]
nums.append(6)   # [1, 2, 3, 4, 5, 6]
nums.insert(0, 0)# [0, 1, 2, 3, 4, 5, 6]
nums.remove(3)   # removes FIRST occurrence of 3
nums.pop()       # removes and returns last element
nums.pop(0)      # removes and returns element at index 0
len(nums)        # length
nums.sort()      # sorts in place (modifies list)
sorted(nums)     # returns NEW sorted list
nums.reverse()   # reverses in place
nums.index(2)    # index of first occurrence of 2
nums.count(2)    # count occurrences of 2
2 in nums        # True if 2 is in list
```

### Tuples (Immutable, Ordered)
```python
t = (1, 2, 3)
t[0]             # 1
t[1:3]           # (2, 3)
# t[0] = 5      # ERROR! Tuples are immutable
len(t)           # 3
1 in t           # True
a, b, c = t      # unpacking: a=1, b=2, c=3
single = (5,)    # single-element tuple NEEDS trailing comma
not_tuple = (5)  # this is just int 5, NOT a tuple!
```

### Sets (Mutable, Unordered, No Duplicates)
```python
s = {1, 2, 3, 3}  # {1, 2, 3} — duplicates removed
s.add(4)           # {1, 2, 3, 4}
s.remove(2)        # {1, 3, 4} — raises error if not found
s.discard(99)      # no error if not found
s1 = {1, 2, 3}
s2 = {2, 3, 4}
s1 | s2            # {1, 2, 3, 4} — union
s1 & s2            # {2, 3} — intersection
s1 - s2            # {1} — difference
s1 ^ s2            # {1, 4} — symmetric difference
```

### Dictionaries (Mutable, Key-Value Pairs)
```python
d = {"name": "Casey", "age": 21}
d["name"]          # "Casey"
d["gpa"] = 3.8     # add new key
d.get("name")      # "Casey" (safe — returns None if missing)
d.get("x", 0)      # 0 (default if missing)
d.keys()           # dict_keys(["name", "age", "gpa"])
d.values()         # dict_values(["Casey", 21, 3.8])
d.items()          # dict_items([("name","Casey"), ("age",21), ("gpa",3.8)])
del d["age"]       # remove key
"name" in d        # True (checks KEYS only)
d.pop("gpa")       # removes and returns value
```

### Data Structure Comparison
| Feature | List | Tuple | Set | Dict |
|---------|------|-------|-----|------|
| Syntax | `[1,2,3]` | `(1,2,3)` | `{1,2,3}` | `{"a":1}` |
| Mutable? | Yes | **No** | Yes | Yes |
| Ordered? | Yes | Yes | **No** | Yes (3.7+) |
| Duplicates? | Yes | Yes | **No** | Keys: No |
| Indexable? | Yes | Yes | **No** | By key |
| Use case | General collection | Fixed data, dict keys | Unique items, membership | Key-value lookup |

---

## Chapter 4 — Branches (If/Else)

### Basic If-Else
```python
if condition:
    # runs if True
elif other_condition:
    # runs if first was False and this is True
else:
    # runs if all above were False
```

### Comparison Operators
| Operator | Meaning | Example |
|----------|---------|---------|
| `==` | Equal to | `x == 5` |
| `!=` | Not equal to | `x != 5` |
| `<` | Less than | `x < 5` |
| `>` | Greater than | `x > 5` |
| `<=` | Less than or equal | `x <= 5` |
| `>=` | Greater than or equal | `x >= 5` |

### Logical Operators
| Operator | Meaning | Example |
|----------|---------|---------|
| `and` | Both must be True | `x > 0 and x < 10` |
| `or` | At least one True | `x < 0 or x > 10` |
| `not` | Inverts True/False | `not (x == 5)` |

**Precedence**: `not` > `and` > `or`

### Chained Comparisons
```python
# Python allows chaining!
if 0 < x < 10:      # same as: x > 0 and x < 10
if 1 <= x <= 100:    # same as: x >= 1 and x <= 100
```

### Membership & Identity Operators
```python
# Membership (in / not in)
"a" in "cat"         # True
3 in [1, 2, 3]       # True
"x" not in "hello"   # True

# Identity (is / is not)
x is None            # True if x is None
x is not None        # True if x is not None
# WARNING: Use == for value comparison, 'is' for identity only!
```

### Conditional Expression (Ternary)
```python
# value_if_true if condition else value_if_false
result = "even" if x % 2 == 0 else "odd"
minimum = a if a < b else b
```

### Order of Evaluation (Full)
1. `()` — Parentheses
2. `**` — Exponentiation
3. `+x`, `-x`, `~x` — Unary
4. `*`, `/`, `//`, `%` — Multiplicative
5. `+`, `-` — Additive
6. `<`, `<=`, `>`, `>=`, `!=`, `==` — Comparisons
7. `not` — Logical NOT
8. `and` — Logical AND
9. `or` — Logical OR

### Common Branching Patterns
```python
# Range check with gap
if x < 0:
    print("negative")
elif 0 <= x <= 10:
    print("low")
elif 11 <= x <= 100:
    print("medium")
else:
    print("high")

# Multiple conditions (checking features)
if has_wifi and has_pool:
    print("luxury")
elif has_wifi or has_pool:
    print("nice")
else:
    print("basic")
```

---

## Chapter 5 — Loops

### While Loop
```python
i = 0
while i < 5:
    print(i)
    i += 1      # MUST update or infinite loop!
# Output: 0 1 2 3 4
```

### For Loop
```python
for i in range(5):        # 0, 1, 2, 3, 4
    print(i)

for i in range(2, 8):     # 2, 3, 4, 5, 6, 7
    print(i)

for i in range(0, 10, 2): # 0, 2, 4, 6, 8 (step of 2)
    print(i)

for i in range(5, 0, -1): # 5, 4, 3, 2, 1 (countdown)
    print(i)

for char in "hello":      # iterate over string characters
    print(char)

for item in [1, 2, 3]:    # iterate over list elements
    print(item)
```

### range() Function
| Call | Produces | Notes |
|------|----------|-------|
| `range(5)` | 0, 1, 2, 3, 4 | Start defaults to 0 |
| `range(2, 5)` | 2, 3, 4 | End is exclusive |
| `range(0, 10, 2)` | 0, 2, 4, 6, 8 | Third arg is step |
| `range(5, 0, -1)` | 5, 4, 3, 2, 1 | Negative step = countdown |
| `range(10, 0, -3)` | 10, 7, 4, 1 | Negative step with gap |

### Nested Loops
```python
for i in range(3):
    for j in range(3):
        print(f"({i},{j})", end=" ")
    print()  # newline after inner loop
# Output:
# (0,0) (0,1) (0,2)
# (1,0) (1,1) (1,2)
# (2,0) (2,1) (2,2)
```

### Break and Continue
```python
# break — exit the loop entirely
for i in range(10):
    if i == 5:
        break       # stops at 5, loop ends
    print(i)        # prints 0 1 2 3 4

# continue — skip this iteration, go to next
for i in range(10):
    if i % 2 == 0:
        continue    # skip even numbers
    print(i)        # prints 1 3 5 7 9
```

### Loop Else
```python
# else block runs if loop completes WITHOUT break
for i in range(5):
    if i == 99:
        break
else:
    print("Loop completed normally!")  # THIS runs

for i in range(5):
    if i == 3:
        break
else:
    print("Loop completed normally!")  # This does NOT run
```

### enumerate() — Get Index AND Value
```python
fruits = ["apple", "banana", "cherry"]

# Without enumerate (ugly way)
for i in range(len(fruits)):
    print(f"{i}: {fruits[i]}")

# With enumerate (Pythonic way)
for idx, fruit in enumerate(fruits):
    print(f"{idx}: {fruit}")

# Start counting from 1
for idx, fruit in enumerate(fruits, start=1):
    print(f"{idx}: {fruit}")
```

### Common Loop Patterns
```python
# Accumulator pattern (sum)
total = 0
for num in [1, 2, 3, 4, 5]:
    total += num
print(total)  # 15

# Finding max/min
nums = [3, 1, 4, 1, 5, 9]
max_val = nums[0]
for num in nums:
    if num > max_val:
        max_val = num

# Counting occurrences
count = 0
for char in "hello":
    if char == "l":
        count += 1  # count = 2

# Building a new list
squares = []
for i in range(5):
    squares.append(i ** 2)  # [0, 1, 4, 9, 16]

# Sentinel loop (until specific input)
while True:
    val = input("Enter value (q to quit): ")
    if val == "q":
        break
    print(f"You entered: {val}")
```

### While vs For: When to Use
| Use `for` when... | Use `while` when... |
|-------------------|---------------------|
| You know how many iterations | Number of iterations unknown |
| Iterating over a sequence | Waiting for a condition |
| Using `range()` | User input loops |
| Processing each item in a collection | Game loops, sentinel values |

---

## Chapter 6 — Functions

### Defining Functions
```python
def greet(name):
    '''Docstring: describes what the function does.'''
    print(f"Hello, {name}!")

greet("Casey")  # Hello, Casey!
```

### Return Values
```python
def add(a, b):
    return a + b    # returns a value

result = add(3, 4)  # result = 7

def nothing():
    print("hi")
    # no return statement → returns None implicitly

x = nothing()       # x is None
```

### Multiple Return Values
```python
def min_max(nums):
    return min(nums), max(nums)  # returns a tuple

lo, hi = min_max([3, 1, 4, 1, 5])  # lo=1, hi=5
```

### Default Parameters
```python
def greet(name, greeting="Hello"):
    print(f"{greeting}, {name}!")

greet("Casey")              # Hello, Casey!
greet("Casey", "Hey")       # Hey, Casey!
```

### Keyword Arguments
```python
def describe(name, age, city):
    print(f"{name}, {age}, from {city}")

describe("Casey", 21, "Pomona")           # positional
describe(city="Pomona", name="Casey", age=21)  # keyword (any order)
describe("Casey", city="Pomona", age=21)  # mixed (positional first!)
```

### Arbitrary Arguments
```python
def total(*args):            # *args = tuple of positional args
    return sum(args)

total(1, 2, 3)               # 6
total(1, 2, 3, 4, 5)         # 15

def info(**kwargs):           # **kwargs = dict of keyword args
    for key, val in kwargs.items():
        print(f"{key}: {val}")

info(name="Casey", age=21)   # name: Casey \n age: 21
```

### Scope of Variables
```python
x = 10              # GLOBAL scope

def my_func():
    y = 5            # LOCAL scope — only exists inside this function
    print(x)         # CAN read global x
    # x = 20         # This creates a NEW local x, doesn't change global!

def change_global():
    global x         # explicitly declare you want to modify global x
    x = 20           # NOW this changes the global x

my_func()
print(x)             # still 10
change_global()
print(x)             # now 20
```

### Scope Rules (LEGB)
| Scope | Description | Example |
|-------|-------------|---------|
| **L**ocal | Inside current function | Variables defined in `def` |
| **E**nclosing | Outer function (for nested functions) | Variables in enclosing `def` |
| **G**lobal | Module level | Variables at top of file |
| **B**uilt-in | Python's built-in names | `print`, `len`, `range` |

Python searches in this order: Local → Enclosing → Global → Built-in

### Functions Are Objects
```python
def square(x):
    return x ** 2

# Functions can be assigned to variables
f = square
print(f(5))    # 25

# Functions can be passed as arguments
def apply(func, value):
    return func(value)

apply(square, 5)   # 25
```

### Common Function Errors
| Error | Why | Fix |
|-------|-----|-----|
| `NameError` | Using variable before defining | Define variable first |
| Returning `None` unexpectedly | Forgot `return` statement | Add `return` |
| Modifying global without `global` | Creates local copy instead | Use `global x` |
| Mutable default argument | `def f(lst=[])` — list is shared! | Use `def f(lst=None)` |
| Arguments in wrong order | Positional before keyword | Put positional args first |

---

## Chapter 7 — String Slicing, Formatting & Methods

### String Slicing (Detailed)
```python
s = "DO NOT LIE!"
#    0123456789 10

s[0:2]     # "DO"
s[0:6]     # "DO NOT"
s[7:10]    # "LIE"
s[:5]      # "DO NO" (from start)
s[7:]      # "LIE!" (to end)
s[-4:]     # "LIE!" (last 4)
s[0:-1]    # "DO NOT LIE" (everything except last)
s[::2]     # "D O I!"  (every 2nd char)
s[::-1]    # "!EIL TON OD" (reversed)
```

### Stride (Step) in Slicing
```python
s = "abcdefgh"
s[::2]     # "aceg"   (every 2nd)
s[1::2]    # "bdfh"   (every 2nd, starting at 1)
s[::-1]    # "hgfedcba" (reverse)
s[5:1:-1]  # "fedc"   (reverse slice)
```

### String Methods
```python
s = "Hello World"

# Case methods
s.upper()          # "HELLO WORLD"
s.lower()          # "hello world"
s.title()          # "Hello World"
s.capitalize()     # "Hello world"
s.swapcase()       # "hELLO wORLD"

# Search methods
s.find("World")    # 6 (index of first occurrence, -1 if not found)
s.index("World")   # 6 (same but raises ValueError if not found)
s.rfind("l")       # 9 (finds LAST occurrence)
s.count("l")       # 3

# Boolean check methods
s.isalpha()        # False (has space)
s.isdigit()        # False
s.isalnum()        # False
s.isspace()        # False
s.isupper()        # False
s.islower()        # False
s.startswith("He") # True
s.endswith("ld")   # True

# Modification methods (return NEW string — strings are immutable!)
s.strip()          # removes leading/trailing whitespace
s.lstrip()         # removes leading whitespace only
s.rstrip()         # removes trailing whitespace only
s.replace("World", "Python")  # "Hello Python"
s.center(20, "-")  # "----Hello World-----"
s.ljust(20, ".")   # "Hello World........."
s.rjust(20, ".")   # ".........Hello World"
s.zfill(20)        # "000000000Hello World"
```

### Split and Join
```python
# split() — string → list
"Hello World".split()          # ["Hello", "World"] (splits on whitespace)
"a,b,c".split(",")             # ["a", "b", "c"] (split on comma)
"2026-02-26".split("-")        # ["2026", "02", "26"]
"a::b::c".split("::")          # ["a", "b", "c"]

# join() — list → string
" ".join(["Hello", "World"])   # "Hello World"
",".join(["a", "b", "c"])      # "a,b,c"
"".join(["P","Y","T","H","O","N"])  # "PYTHON"
"-".join(["2026", "02", "26"]) # "2026-02-26"

# GOTCHA: join() requires ALL elements to be strings!
" ".join(["Age:", 25])         # TypeError!
" ".join(["Age:", str(25)])    # "Age: 25" ← convert first
```

### Traversing Lists (3 Ways)
```python
numbers = [10, 20, 30, 40, 50]

# Option 1: by index
for i in range(len(numbers)):
    print(numbers[i])

# Option 2: by value (simple and clean)
for num in numbers:
    print(num)

# Option 3: enumerate (BOTH index and value)
for idx, num in enumerate(numbers):
    print(f"{idx}: {num}")

for idx, num in enumerate(numbers, start=1):
    print(f"Position {idx} → {num}")
```

---

# PART 2: CHEAT SHEET TABLES

---

## Operators Cheat Sheet

| Operator | Name | Example | Result | Notes |
|----------|------|---------|--------|-------|
| `+` | Addition | `3 + 2` | `5` | Also concatenation: `"a" + "b"` → `"ab"` |
| `-` | Subtraction | `3 - 2` | `1` | |
| `*` | Multiplication | `3 * 2` | `6` | Also repetition: `"ab" * 3` → `"ababab"` |
| `/` | True Division | `7 / 2` | `3.5` | Always returns float |
| `//` | Floor Division | `7 // 2` | `3` | Rounds DOWN (toward -∞) |
| `%` | Modulo | `7 % 3` | `1` | Remainder after division |
| `**` | Exponentiation | `2 ** 3` | `8` | Right-to-left: `2**3**2` = `2**9` = 512 |
| `==` | Equal | `5 == 5` | `True` | Value comparison |
| `!=` | Not equal | `5 != 3` | `True` | |
| `<` | Less than | `3 < 5` | `True` | |
| `>` | Greater than | `5 > 3` | `True` | |
| `<=` | Less or equal | `5 <= 5` | `True` | |
| `>=` | Greater or equal | `5 >= 5` | `True` | |
| `and` | Logical AND | `True and False` | `False` | Short-circuits |
| `or` | Logical OR | `True or False` | `True` | Short-circuits |
| `not` | Logical NOT | `not True` | `False` | |
| `in` | Membership | `"a" in "cat"` | `True` | Works on str, list, tuple, set, dict |
| `not in` | Non-membership | `"x" not in "cat"` | `True` | |
| `is` | Identity | `x is None` | | Checks if SAME object |
| `is not` | Not identity | `x is not None` | | |

## String Methods Cheat Sheet

| Method | What It Does | Example | Result |
|--------|-------------|---------|--------|
| `upper()` | All uppercase | `"hi".upper()` | `"HI"` |
| `lower()` | All lowercase | `"HI".lower()` | `"hi"` |
| `title()` | Title Case | `"hi there".title()` | `"Hi There"` |
| `capitalize()` | First letter caps | `"hi there".capitalize()` | `"Hi there"` |
| `swapcase()` | Swap case | `"Hi".swapcase()` | `"hI"` |
| `strip()` | Remove whitespace both ends | `" hi ".strip()` | `"hi"` |
| `lstrip()` | Remove left whitespace | `" hi ".lstrip()` | `"hi "` |
| `rstrip()` | Remove right whitespace | `" hi ".rstrip()` | `" hi"` |
| `find(x)` | Index of first x (-1 if none) | `"hello".find("l")` | `2` |
| `rfind(x)` | Index of last x | `"hello".rfind("l")` | `3` |
| `index(x)` | Like find but raises error | `"hello".index("l")` | `2` |
| `count(x)` | Count occurrences | `"hello".count("l")` | `2` |
| `replace(a,b)` | Replace a with b | `"hi".replace("i","ey")` | `"hey"` |
| `split(x)` | Split into list | `"a,b".split(",")` | `["a","b"]` |
| `join(list)` | Join list into string | `"-".join(["a","b"])` | `"a-b"` |
| `startswith(x)` | Starts with x? | `"hello".startswith("he")` | `True` |
| `endswith(x)` | Ends with x? | `"hello".endswith("lo")` | `True` |
| `isdigit()` | All digits? | `"123".isdigit()` | `True` |
| `isalpha()` | All letters? | `"abc".isalpha()` | `True` |
| `isalnum()` | All alphanumeric? | `"abc123".isalnum()` | `True` |
| `isspace()` | All whitespace? | `"  ".isspace()` | `True` |
| `center(w,c)` | Center in width w | `"hi".center(10,"-")` | `"----hi----"` |
| `ljust(w,c)` | Left-justify | `"hi".ljust(10,".")` | `"hi........"` |
| `rjust(w,c)` | Right-justify | `"hi".rjust(10,".")` | `"........hi"` |
| `zfill(w)` | Zero-pad | `"42".zfill(5)` | `"00042"` |

## List/Tuple/Set/Dict Methods Cheat Sheet

### List Methods
| Method | What It Does | Example | Result |
|--------|-------------|---------|--------|
| `append(x)` | Add to end | `[1,2].append(3)` | `[1,2,3]` |
| `insert(i,x)` | Insert at index | `[1,3].insert(1,2)` | `[1,2,3]` |
| `remove(x)` | Remove first x | `[1,2,2].remove(2)` | `[1,2]` |
| `pop()` | Remove & return last | `[1,2,3].pop()` | returns `3` |
| `pop(i)` | Remove & return at i | `[1,2,3].pop(0)` | returns `1` |
| `sort()` | Sort in place | `[3,1,2].sort()` | `[1,2,3]` |
| `reverse()` | Reverse in place | `[1,2,3].reverse()` | `[3,2,1]` |
| `index(x)` | Index of first x | `[1,2,3].index(2)` | `1` |
| `count(x)` | Count of x | `[1,2,2].count(2)` | `2` |
| `extend(lst)` | Add all from another list | `[1].extend([2,3])` | `[1,2,3]` |
| `copy()` | Shallow copy | `[1,2].copy()` | `[1,2]` |
| `clear()` | Remove all | `[1,2].clear()` | `[]` |

### Set Methods
| Method | What It Does | Example |
|--------|-------------|---------|
| `add(x)` | Add element | `{1,2}.add(3)` → `{1,2,3}` |
| `remove(x)` | Remove (error if missing) | `{1,2}.remove(1)` → `{2}` |
| `discard(x)` | Remove (no error) | `{1,2}.discard(99)` → `{1,2}` |
| `union(s)` or `\|` | Combine all | `{1,2} \| {2,3}` → `{1,2,3}` |
| `intersection(s)` or `&` | Common elements | `{1,2} & {2,3}` → `{2}` |
| `difference(s)` or `-` | In first, not second | `{1,2} - {2,3}` → `{1}` |
| `symmetric_difference(s)` or `^` | In one but not both | `{1,2} ^ {2,3}` → `{1,3}` |

### Dict Methods
| Method | What It Does | Example |
|--------|-------------|---------|
| `get(key, default)` | Safe access | `d.get("x", 0)` → `0` |
| `keys()` | All keys | `d.keys()` |
| `values()` | All values | `d.values()` |
| `items()` | Key-value pairs | `d.items()` |
| `pop(key)` | Remove & return value | `d.pop("name")` |
| `update(d2)` | Merge another dict | `d.update({"a":1})` |
| `setdefault(k,v)` | Get or set if missing | `d.setdefault("x", 0)` |
| `in` | Check if key exists | `"name" in d` → `True` |

## Format Specifiers Cheat Sheet

| Format | Meaning | `f"{value:FORMAT}"` | Result |
|--------|---------|---------------------|--------|
| `.2f` | 2 decimal places | `f"{3.14159:.2f}"` | `3.14` |
| `.0f` | No decimal places | `f"{3.14:.0f}"` | `3` |
| `10d` | Int, width 10 | `f"{42:10d}"` | `"        42"` |
| `05d` | Int, zero-padded width 5 | `f"{42:05d}"` | `"00042"` |
| `>10` | Right-align, width 10 | `f"{'hi':>10}"` | `"        hi"` |
| `<10` | Left-align, width 10 | `f"{'hi':<10}"` | `"hi        "` |
| `^10` | Center, width 10 | `f"{'hi':^10}"` | `"    hi    "` |
| `10.2f` | Float, width 10, 2 dec | `f"{3.14:10.2f}"` | `"      3.14"` |
| `.2e` | Scientific, 2 decimals | `f"{1234:.2e}"` | `"1.23e+03"` |
| `.1%` | Percentage, 1 decimal | `f"{0.856:.1%}"` | `"85.6%"` |
| `,` | Thousands separator | `f"{1000000:,}"` | `"1,000,000"` |

## Loop Patterns Cheat Sheet

| Pattern | Code | Use Case |
|---------|------|----------|
| Count-controlled | `for i in range(n):` | Known iterations |
| Collection iteration | `for item in collection:` | Process each element |
| Index-based | `for i in range(len(lst)):` | Need index access |
| enumerate | `for i, val in enumerate(lst):` | Need both index and value |
| Sentinel | `while True: ... if cond: break` | Until specific condition |
| Accumulator | `total = 0; for x in lst: total += x` | Sum/count/build |
| Filter | `for x in lst: if cond: new.append(x)` | Select matching items |
| Nested | `for i in ...: for j in ...:` | 2D grids, combinations |
| Countdown | `for i in range(n, 0, -1):` | Reverse counting |
| Skip (continue) | `if cond: continue` | Skip certain iterations |
| Early exit (break) | `if cond: break` | Stop when found |

---

# PART 3: COMMON MISTAKES & GOTCHAS

---

| # | Mistake | Wrong | Right | Why |
|---|---------|-------|-------|-----|
| 1 | Forgetting `int()` on `input()` | `age = input("Age: ")` then `age + 1` → TypeError | `age = int(input("Age: "))` | `input()` always returns a string |
| 2 | Using `=` instead of `==` | `if x = 5:` → SyntaxError | `if x == 5:` | `=` is assignment, `==` is comparison |
| 3 | Off-by-one in `range()` | `range(1, 5)` expecting 5 | `range(1, 6)` to include 5 | End is exclusive |
| 4 | Off-by-one in slicing | `s[0:3]` expecting 4 chars | `s[0:4]` | End index is exclusive |
| 5 | Modifying list while iterating | `for x in lst: lst.remove(x)` | Use list comprehension or copy | Skips elements |
| 6 | `int("3.5")` | Crashes with ValueError | `int(float("3.5"))` | Can't convert float-string directly |
| 7 | `int()` rounds down, not nearest | `int(3.9)` → `3` | `round(3.9)` → `4` | `int()` truncates toward zero |
| 8 | Infinite while loop | Forgetting to update counter | Always update loop variable | `i += 1` inside the loop |
| 9 | Single-element tuple | `t = (5)` → just `int` 5 | `t = (5,)` | Need trailing comma |
| 10 | `is` vs `==` | `x is 5` (unreliable) | `x == 5` | `is` checks identity, `==` checks value |
| 11 | Mutable default args | `def f(lst=[]):` | `def f(lst=None): if lst is None: lst = []` | Default list is shared across calls |
| 12 | `join()` with non-strings | `" ".join(["a", 1])` → TypeError | `" ".join(["a", str(1)])` | All elements must be strings |
| 13 | `sort()` vs `sorted()` | `y = x.sort()` → `y` is None | `y = sorted(x)` | `.sort()` modifies in place, returns None |
| 14 | String immutability | `s[0] = "X"` → TypeError | `s = "X" + s[1:]` | Strings cannot be changed in place |
| 15 | Scope confusion | Assigning to global inside function | Use `global x` keyword | Assignment creates local variable |
| 16 | `remove()` on set | `s.remove(99)` → KeyError | `s.discard(99)` | `discard` doesn't error on missing |
| 17 | Empty string/list is falsy | `if my_str:` skips empty string | Intentional — useful for checks | `""`, `[]`, `0`, `None` are all False |
| 18 | `-4 ** 2` = `-16` | Expected `16` | `(-4) ** 2` = `16` | `**` has higher precedence than unary `-` |
| 19 | `//` rounds toward -∞ | `-7 // 2` = `-4` (not -3) | It's floor division, not truncation | Always rounds DOWN |
| 20 | Forgetting `return` | Function returns `None` | Add explicit `return value` | No return = returns None |

---

# PART 4: SYNTAX QUICK-REFERENCE CARD

---

```
VARIABLES          x = 5                    name = "Casey"
MULTIPLE ASSIGN    x, y = 1, 2              a, b = b, a  (swap)
INPUT              x = input("prompt: ")    n = int(input("num: "))
OUTPUT             print(x)                 print(f"x = {x}")
                   print(x, end="")         print(x, y, sep=", ")

IF/ELIF/ELSE       if x > 0:               elif x == 0:              else:
TERNARY            result = "yes" if x else "no"

FOR LOOP           for i in range(5):       for item in mylist:
                   for i, v in enumerate(lst):
WHILE LOOP         while condition:         while True: ... break
BREAK/CONTINUE     break (exit loop)        continue (skip iteration)

FUNCTION DEF       def func(param):         def func(a, b=10):
                       return value              return a + b
FUNCTION CALL      func(5)                  func(a=1, b=2)

STRING SLICE       s[start:end]             s[start:end:step]
                   s[:5]  s[5:]  s[::-1]

LIST CREATE        lst = [1, 2, 3]          lst = list(range(5))
LIST ADD           lst.append(x)            lst.insert(i, x)
LIST REMOVE        lst.remove(x)            lst.pop()    lst.pop(i)
LIST SORT          lst.sort()               sorted(lst)

DICT CREATE        d = {"key": "val"}       d = dict()
DICT ACCESS        d["key"]                 d.get("key", default)
DICT ADD           d["new_key"] = value
DICT LOOP          for k, v in d.items():

SET CREATE         s = {1, 2, 3}            s = set()
SET OPS            s1 | s2 (union)          s1 & s2 (intersection)
                   s1 - s2 (difference)     s1 ^ s2 (symmetric diff)

TYPE CHECK         type(x)                  isinstance(x, int)
TYPE CONVERT       int(x)  float(x)  str(x)  bool(x)  list(x)  tuple(x)  set(x)

IMPORT             import math              from math import sqrt
                   import random            from random import randint
```

---

# PART 5: PRACTICE PATTERNS (What Professors Love to Test)

---

### Pattern 1: Palindrome Check
```python
def is_palindrome(s):
    s = s.lower().replace(" ", "")
    return s == s[::-1]
```

### Pattern 2: Two-Pointer / Reverse
```python
def reverse_string(s):
    return s[::-1]

def reverse_list(lst):
    return lst[::-1]
```

### Pattern 3: 2D List Flatten
```python
def flatten(matrix):
    result = []
    for row in matrix:
        for item in row:
            result.append(item)
    return result
# flatten([[1,2],[3,4],[5,6]]) → [1,2,3,4,5,6]
```

### Pattern 4: Find Sequence in Array
```python
def array123(nums):
    for i in range(len(nums) - 2):
        if nums[i:i+3] == [1, 2, 3]:
            return True
    return False
```

### Pattern 5: Count Characters
```python
def char_count(s):
    counts = {}
    for char in s:
        counts[char] = counts.get(char, 0) + 1
    return counts
```

### Pattern 6: Scope Tracing
```python
x = 10
def foo():
    x = 20       # local x, doesn't change global
    print(x)     # 20

foo()
print(x)         # 10 (global unchanged)
```

### Pattern 7: Input Validation Loop
```python
while True:
    try:
        n = int(input("Enter a number: "))
        break
    except ValueError:
        print("That's not a number!")
```

### Pattern 8: Accumulator with Condition
```python
def sum_evens(nums):
    total = 0
    for n in nums:
        if n % 2 == 0:
            total += n
    return total
```

### Pattern 9: String Manipulation
```python
def first_last(s):
    if len(s) <= 1:
        return s
    return s[-1] + s[1:-1] + s[0]
# first_last("hello") → "oellh"
```

### Pattern 10: Exact Change / Breaking Down Numbers
```python
def exact_change(cents):
    quarters = cents // 25
    cents %= 25
    dimes = cents // 10
    cents %= 10
    nickels = cents // 5
    cents %= 5
    pennies = cents
    return quarters, dimes, nickels, pennies
```

---

*Study guide generated by Dan — March 12, 2026*
*Sources: ZyBooks Ch 1-7 (109 sections), Lecture Slides Lec_02 through Lec_07 (376 pages)*

"""
pp(qwertyuiop)
````
