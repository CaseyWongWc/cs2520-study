```python
from IPython import get_ipython
from IPython.display import Markdown, display
import ast

def _bare_string_to_markdown(tree):
    class _T(ast.NodeTransformer):
        def visit_Expr(self, node):
            if isinstance(node.value, ast.Constant) and isinstance(node.value.value, str):
                return ast.copy_location(
                    ast.Expr(
                        value=ast.Call(
                            func=ast.Name(id='display', ctx=ast.Load()),
                            args=[
                                ast.Call(
                                    func=ast.Name(id='Markdown', ctx=ast.Load()),
                                    args=[node.value],
                                    keywords=[],
                                )
                            ],
                            keywords=[],
                        )
                    ),
                    node,
                )
            return node

    return _T().visit(tree)

ip = get_ipython()
# add once
if _bare_string_to_markdown not in ip.ast_transformers:
    ip.ast_transformers.append(_bare_string_to_markdown)

# test:
"""## header\nthis should render as markdown"""
```

```python
from IPython.display import Markdown, display
import inspect

def display_md(text: str):
    """Render a plain string as Markdown."""
    display(Markdown(text))

def render_docstring(obj):
    """Render the docstring of any object as Markdown."""
    ds = inspect.getdoc(obj)
    if not ds:
        raise ValueError(f"No docstring found for {obj!r}")
    display(Markdown(ds))

# optional: shortcut for quoted strings (if you like explicit call)
def m(s: str):
    display_md(s)
```
