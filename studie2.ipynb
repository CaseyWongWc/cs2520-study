### Two-Pointer / Palindrome (QUIZ TOPIC!)

```python
# Palindrome check — TWO-POINTER technique
def is_palindrome(s):
    left = 0
    right = len(s) - 1
    while left < right:
        if s[left] != s[right]:
            return False
        left += 1
        right -= 1
    return True

# Quick palindrome check (slicing)
def is_palindrome_easy(s):
    return s == s[::-1]

# Reverse a list
a = [1, 2, 3, 4]
t1 = list(a)
t2 = list(a)
t2.reverse()             # in-place reverse
# OR
reversed_a = a[::-1]     # new reversed copy
```

## Ch 5: Loops

```python
# While loop
i = 0
while i < 5:
    print(i)
    i += 1

# For loop with range
for i in range(5):         # 0, 1, 2, 3, 4
for i in range(2, 8):      # 2, 3, 4, 5, 6, 7
for i in range(0, 10, 2):  # 0, 2, 4, 6, 8 (step=2)
for i in range(5, 0, -1):  # 5, 4, 3, 2, 1 (countdown)

# Loop through string
for char in "hello":
    print(char)

# Loop through list
for item in [1, 2, 3]:
    print(item)

# Loop with index
for i, item in enumerate(['a', 'b', 'c']):
    print(i, item)        # 0 a, 1 b, 2 c

# break = exit loop immediately
# continue = skip to next iteration

# LOOP ELSE — runs ONLY if loop finishes WITHOUT break
for x in [2, 4, 6]:
    if x == 5:
        break
else:
    print("No break!")       # PRINTS (loop finished normally)

for x in [2, 5, 6]:
    if x == 5:
        break
else:
    print("No break!")       # DOES NOT PRINT (break was hit)

# Set: discard() vs remove()
s = {1, 2, 3}
s.remove(99)                 # KeyError! CRASHES
s.discard(99)                # No error (safe version)

# Nested loops
for i in range(3):
    for j in range(3):
        print(i, j)

# Sum pattern
total = 0
for x in [1, 2, 3, 4, 5]:
    total += x

# Count pattern
count = 0
for x in [1, 2, 3, 4, 5]:
    if x > 2:
        count += 1

# Min/Max pattern
nums = [3, 1, 4, 1, 5]
min_val = nums[0]
for x in nums:
    if x < min_val:
        min_val = x
```

## Ch 6: Functions

```python
# Define a function
def greet(name):
    return f"Hello, {name}!"

# Call it
message = greet("Casey")  # → "Hello, Casey!"

# Default parameters
def greet(name, greeting="Hello"):
    return f"{greeting}, {name}!"

greet("Casey")             # → "Hello, Casey!"
greet("Casey", "Hi")       # → "Hi, Casey!"

# Multiple return values
def min_max(nums):
    return min(nums), max(nums)

lo, hi = min_max([3, 1, 4])  # lo=1, hi=4

# Scope: variables inside function are LOCAL
x = 10             # global
def foo():
    x = 5          # local x (different from global x)
    return x
foo()               # → 5
print(x)            # → 10 (global unchanged)

# None — function without return gives None
def say_hi():
    print("Hi")

result = say_hi()   # prints "Hi", result is None

# Lists are passed by REFERENCE (mutable)
def add_item(lst):
    lst.append(99)

my_list = [1, 2, 3]
add_item(my_list)
print(my_list)      # → [1, 2, 3, 99] — MODIFIED!

# Strings/ints are passed by VALUE (immutable)
def change(x):
    x = 999

val = 5
change(val)
print(val)           # → 5 — NOT modified
```

### 2D Lists (Flatten — quiz topic!)
```python
matrix = [[1, 2], [3, 4], [5, 6]]

# Access element
matrix[0][1]           # → 2 (row 0, col 1)

# Flatten (2D → 1D)
def flatten(matrix):
    result = []
    for row in matrix:
        for item in row:
            result.append(item)
    return result

flatten([[1,2],[3,4]])  # → [1, 2, 3, 4]

# List comprehension version
flat = [item for row in matrix for item in row]
```

## Gotchas to Watch For

| Trap | What happens |
|------|-------------|
| `int(3.7)` | → 3 (truncates, doesn't round!) |
| `round(2.5)` | → 2 (banker's rounding in Python 3!) |
| `"5" + 3` | TypeError! (can't add str + int) |
| `s[0] = 'x'` | TypeError! (strings are immutable) |
| `range(5)` | 0-4, NOT 0-5 |
| `lst = [1,2]; lst2 = lst` | Both point to SAME list! |
| `def f(x=[]): x.append(1)` | Mutable default — shared across calls! |
| `return` vs `print` | `return` gives value back, `print` just displays |
| `-4 ** 2` | → -16 (not 16! `**` binds tighter than unary `-`) |
| `-7 // 2` | → -4 (floor division rounds toward -∞, not toward 0) |
| `int("3.5")` | ValueError! Must do `int(float("3.5"))` |
| `join()` with non-strings | `" ".join(["a", 1])` → TypeError! |
| `sort()` returns None | `y = lst.sort()` → y is None, use `y = sorted(lst)` |

## Ch 7: Sequences — Strings, Lists, split/join, enumerate

```python
# split() — break string into list
"Hello World".split()          # → ['Hello', 'World']
"a,b,c".split(",")             # → ['a', 'b', 'c']
"2026-03-12".split("-")        # → ['2026', '03', '12']

# join() — combine list into string (ALL elements must be strings!)
" ".join(["Hello", "World"])   # → 'Hello World'
",".join(["a", "b", "c"])      # → 'a,b,c'
"".join(["P","Y","T","H"])     # → 'PYTH'

# enumerate() — get index AND value
fruits = ["apple", "banana", "cherry"]
for idx, fruit in enumerate(fruits):
    print(f"{idx}: {fruit}")   # 0: apple, 1: banana, 2: cherry

for idx, fruit in enumerate(fruits, start=1):
    print(f"{idx}: {fruit}")   # 1: apple, 2: banana, 3: cherry

# 3 ways to traverse a list
# 1. By index:     for i in range(len(lst)):   lst[i]
# 2. By value:     for item in lst:            item
# 3. enumerate:    for i, item in enumerate(lst): i, item

# Stride (step) in slicing
s = "abcdefgh"
s[::2]                         # → 'aceg'  (every 2nd char)
s[1::2]                        # → 'bdfh'  (every 2nd, start at 1)
s[::-1]                        # → 'hgfedcba' (reverse)

# String formatting with f-strings
name = "Casey"
gpa = 3.856
print(f"GPA: {gpa:.2f}")      # → GPA: 3.86
print(f"{'hello':>10}")        # → '     hello' (right-align)
print(f"{'hello':<10}")        # → 'hello     ' (left-align)
print(f"{'hello':^10}")        # → '  hello   ' (center)
print(f"{42:05d}")             # → '00042' (zero-padded)
print(f"{0.85:.1%}")           # → '85.0%' (percentage)
print(f"{1000000:,}")          # → '1,000,000' (thousands sep)

# array123 pattern (find [1,2,3] subsequence)
def array123(nums):
    for i in range(len(nums) - 2):
        if nums[i:i+3] == [1, 2, 3]:
            return True
    return False
```

### TRAP 1: Negative Floor Division (floors toward -∞)
```python
-10 // 3    # → -4   NOT -3!  (floor of -3.33 = -4)
-7 // 2     # → -4   NOT -3!  (floor of -3.5 = -4)
10 // 3     # → 3    (floor of 3.33 = 3, positive works as expected)
```

### TRAP 2: `total += i` adds the VALUE, not the count
```python
total = 0
for i in range(1, 11):
    if i % 3 == 0:
        total += i       # adds 3, then 6, then 9
print(total)             # → 18  NOT 6 (that would be counting)
```

### TRAP 3: `.find()` returns INDEX of substring start
```python
s = "Python Programming"
s.find('gram')    # → 10  (where 'g' of 'gram' is, NOT where word starts)
s.find('Pro')     # → 7   (where 'P' of 'Pro' is)
s.find('xyz')     # → -1  (not found)
# COUNT characters from 0 to find the index!
```

### TRAP 4: `center()`/`ljust()`/`rjust()` DON'T change case
```python
x = "hello"
x.center(11, '-')  # → '---hello---'  NOT '---Hello---'
x.ljust(10, '.')   # → 'hello.....'  NOT 'Hello.....'
# These ONLY add padding — they never capitalize!
```

### TRAP 5: `%` format multiplies by 100
```python
f"{0.856:.0%}"   # → '86%'   (0.856 × 100 = 85.6, round to 0 dec = 86)
f"{0.856:.1%}"   # → '85.6%'
# The % format AUTOMATICALLY multiplies by 100!
```

### TRAP 6: `05d` = zero-padded integer
```python
f"|{21:05d}|"    # → '|00021|'   NOT '|   21|'
f"|{21:5d}|"     # → '|   21|'   (space-padded, no zero)
```

### TRAP 7: `sorted()` vs `.sort()` — returns matter
```python
words = ['banana', 'apple']
result = sorted(words)      # NEW list, words unchanged
words.sort()                # IN-PLACE, returns None
words.sort(reverse=True)    # reverse in-place
```

### TRAP 8: `.lower()` for word_count!
```python
def word_count(text):
    d = {}
    for w in text.lower().split():  # .lower() FIRST!
        if w not in d:
            d[w] = 1
        else:
            d[w] += 1
    return d
```

## Caesar Cipher (HARDEST EXAM PATTERN)

```python
def caesar(text, shift):
    result = ""
    for ch in text:
        if ch.isalpha():
            base = ord('a') if ch.islower() else ord('A')
            result += chr((ord(ch) - base + shift) % 26 + base)
        else:
            result += ch
    return result

# Step-by-step breakdown (TI-BASIC style):
# z1 = ord(ch)            ← letter to number (e.g. 'x' → 120)
# z2 = z1 - base          ← position 0-25 (e.g. 120-97 = 23)
# z3 = (z2 + shift) % 26  ← shift + wrap (e.g. (23+3)%26 = 0)
# z4 = z3 + base          ← back to ASCII (e.g. 0+97 = 97)
# z5 = chr(z4)            ← number to letter (e.g. 97 → 'a')
```

## 8 Functions to Write from Memory

```python
# 1. Sum of even numbers
def sum_evens(n):
    total = 0
    for i in range(2, n + 1, 2):  # n+1 for inclusive!
        total += i
    return total

# 2. Count vowels
def count_vowels(text):
    count = 0
    for ch in text.lower():
        if ch in 'aeiou':
            count += 1
    return count

# 3. Reverse a string
def reverse_str(s):
    result = ''
    for ch in s:
        result = ch + result    # prepend each char
    return result
# OR simply: return s[::-1]

# 4. Word count (dict accumulator)
def word_count(text):
    d = {}
    for w in text.lower().split():
        if w not in d:
            d[w] = 1
        else:
            d[w] += 1
    return d

# 5. is_palindrome (case-insensitive, ignore spaces)
def is_palindrome(s):
    s = s.lower().replace(' ', '')
    return s == s[::-1]

# 6. flatten_2d
def flatten_2d(matrix):
    result = []
    for row in matrix:
        for item in row:
            result.append(item)
    return result

# 7. remove_duplicates (keep order)
def remove_duplicates(lst):
    seen = []
    for item in lst:
        if item not in seen:
            seen.append(item)
    return seen

# 8. char_frequency (ignore spaces)
def char_frequency(s):
    d = {}
    for ch in s:
        if ch != ' ':
            if ch not in d:
                d[ch] = 1
            else:
                d[ch] += 1
    return d
```
