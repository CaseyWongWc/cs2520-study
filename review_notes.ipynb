## 🔤 Section 2: String Methods
You practiced `.lower()`, `.upper()`, and `.replace()` — these are called **methods** (functions attached to objects).

| Method | What it does | Example |
|--------|-------------|---------|
| `.lower()` | Makes everything lowercase | `'HELLO'.lower()` → `'hello'` |
| `.upper()` | Makes everything uppercase | `'hello'.upper()` → `'HELLO'` |
| `.replace(a, b)` | Swaps `a` for `b` | `'hi yo'.replace(' ','')` → `'hiyo'` |

````python
s = 'Thello the Great Thane is Very Great'  # your string from practice!

print('Original  :', s)
print('Lower     :', s.lower())
print('Upper     :', s.upper())
print('No spaces :', s.replace(' ', ''))
print('Lower+nospace:', s.lower().replace(' ', ''))  # chaining methods!
````

## 🔧 Section 3: Functions (def & return)
You built two functions in your session: `e(length)` and `s()`. Here's the pattern:

```python
def function_name(parameters):
    # body must be INDENTED (4 spaces)
    return result
```

⚠️ **Most common mistake you made:** forgetting `return` or indentation!

````python
# e(length) — generates a random string of letters
def e(length):
    chars = string.ascii_letters  # a-zA-Z
    return ''.join(random.choice(chars) for _ in range(length))

# s() — calls e(5) and returns it
def s():
    return e(5)

# Try them out!
print('e(5)        :', e(5))
print('s()         :', s())
print('s().lower() :', s().lower())
print('s().upper() :', s().upper())
````

## 🧠 Section 7: Dynamic Typing vs Static Typing
(From your lecture slides!)

| | Dynamic (Python) | Static (C++, Java) |
|---|---|---|
| Type checked | At **runtime** | At **compile time** |
| Declare type? | ❌ No | ✅ Yes (`int x = 5`) |
| Variable can change type? | ✅ Yes | ❌ No |
| Speed | Slower | Faster |
| Example | `x = 5` then `x = 'hi'` ✅ | `int x = 5` then `x = 'hi'` ❌ |

````python
# Python lets you change a variable's type freely — that's dynamic typing!
x = 5
print(type(x), x)   # <class 'int'>

x = 'hello'
print(type(x), x)   # <class 'str'>  -- no error!

x = [1, 2, 3]
print(type(x), x)   # <class 'list'> -- still no error!

# In C++ this would crash. In Python it's totally fine.
````

## 🔒 Section 8: The Global Interpreter Lock (GIL)
(From your lecture slides!)

The **GIL** is a mutex (lock) inside CPython that:
- Only allows **one Python thread to execute at a time**
- Protects Python's memory management from concurrent access bugs
- Means Python threads don't truly run in parallel for CPU-heavy tasks

**Analogy:** Imagine a kitchen with one knife 🔪. Even if 3 chefs (threads) are working, only one can use the knife at a time.

**Workaround:** Use `multiprocessing` instead of `threading` for CPU-heavy work — each process gets its own GIL!
