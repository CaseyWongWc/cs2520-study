## 🖨️ Section 0: Python Basic Output — print()
*(From your lecture slides!)*

The `print()` function displays output to the user. It has two secret superpowers: `sep` and `end`!

| Parameter | What it does | Default |
|-----------|-------------|--------|
| `sep` | What goes **between** multiple values | `' '` (a space) |
| `end` | What goes at the **end** of the line | `'\n'` (newline) |

**Basic syntax:**
```python
print(value1, value2, ..., sep='separator', end='ending')
```

````python
# Basic print — default sep is a space between values
print('Today', 'is', 'Monday.')          # → Today is Monday.
print('I', 'like', 'string', 'beans.')   # → I like string beans.
````

````python
# sep= changes what goes BETWEEN values
print('Today', 'is', 'Monday.', sep='-')      # → Today-is-Monday.
print('Today', 'is', 'Monday.', sep='...')    # → Today...is...Monday.
print('Today', 'is', 'Monday.', sep='')       # → TodayisMonday.  (no space!)
````

````python
# end= changes what goes at the END of the line
# Default end='\n' means each print goes on a new line
# Change it to '' to stay on the same line!

print('Today', end=' ')    # stays on same line!
print('is', end=' ')       # still same line!
print('Monday.')           # NOW it goes to next line
# → Today is Monday.
````

````python
# Combining sep AND end together!
print('Today', 'is', 'Monday.', sep=', ', end='!!\n')

# You already used end='' in your count_vowels() — that's how h[e]ll[o] worked!
for char in 'Hello World'.lower():
    if char in 'aeiou':
        print('[' + char + ']', end='')
    else:
        print(char, end='')
print()  # bare print() = just a newline
````

## 🔤 Section 2: String Methods
| Method | What it does | Example |
|--------|-------------|--------|
| `.lower()` | Makes everything lowercase | `'HELLO'.lower()` → `'hello'` |
| `.upper()` | Makes everything uppercase | `'hello'.upper()` → `'HELLO'` |
| `.replace(a, b)` | Swaps `a` for `b` | `'hi yo'.replace(' ','')` → `'hiyo'` |

````python
s = 'Thello the Great Thane is Very Great'
print('Original  :', s)
print('Lower     :', s.lower())
print('Upper     :', s.upper())
print('No spaces :', s.replace(' ', ''))
print('Chained   :', s.lower().replace(' ', ''))
````

## 🔧 Section 3: Functions (def & return)
```python
def function_name(parameters):
    # body must be INDENTED (4 spaces)
    return result
```
⚠️ Most common mistake: forgetting `return` or indentation!

````python
def e(length):
    chars = string.ascii_letters
    return ''.join(random.choice(chars) for _ in range(length))

def s():
    return e(5)

print('e(5):', e(5))
print('s():', s())
print('s().lower():', s().lower())
````

## ✏️ Section 5: YOUR CHALLENGE — count_vowels()
**You already solved this! 🎉 Your creative visual version is below — it uses `end=''` to stay on one line!**

````python
# Clean version
def count_vowels(text):
    count = 0
    for char in text.lower():
        if char in 'aeiou':
            count += 1
    return count

print(count_vowels('Hello World'))  # → 3
print(count_vowels('aeiou'))        # → 5
````

````python
# YOUR CREATIVE VISUAL VERSION 🎨 (uses end='' trick from Section 0!)
def count_vowels_visual(text):
    count = 0
    for char in text.lower():
        if char in 'aeiou':
            print('[' + char + ']', end='')
            count += 1
        else:
            print(char, end='')
    print()
    return count

print(count_vowels_visual('Hello World'))  # h[e]ll[o] w[o]rld → 3
````

## 💡 Section 6: Common Errors
| Error | What caused it | Fix |
|-------|---------------|-----|
| `NameError: 'string' is not defined` | Used before importing | Add `import string` at top |
| `IndentationError` | Function body not indented | Use 4 spaces after `def` |
| `TypeError: 'str' object is not callable` | Overwrote a function name with a string | Rename variable or restart |
| `SyntaxError` | Pasted markdown backtick fences | Paste plain Python only |

## 🧠 Section 7: Dynamic Typing vs Static Typing
| | Dynamic (Python) | Static (C++, Java) |
|---|---|---|
| Type checked | At **runtime** | At **compile time** |
| Declare type? | ❌ No | ✅ Yes (`int x = 5`) |
| Variable can change type? | ✅ Yes | ❌ No |
| Speed | Slower | Faster |

````python
x = 5
print(type(x), x)
x = 'hello'
print(type(x), x)
x = [1, 2, 3]
print(type(x), x)
# In C++ this would crash. In Python it's fine!
````

## 🔒 Section 8: The Global Interpreter Lock (GIL)
The **GIL** is a mutex (lock) inside CPython that:
- Only allows **one Python thread to execute at a time**
- Protects Python's memory management from concurrent access bugs
- Means Python threads don't truly run in parallel for CPU-heavy tasks

**Analogy:** Imagine a kitchen with one knife 🔪. Even if 3 chefs (threads) are working, only one can use the knife at a time.

**Workaround:** Use `multiprocessing` instead of `threading` for CPU-heavy work!
