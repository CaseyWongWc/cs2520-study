```python
def stringfromcharacters(start, end):
    accumulator = ""
    for x in range(ord(start), ord(end)+1):
        accumulator += chr(x)
    return accumulator
stringfromcharacters('a', 'z')
stringfromcharacters('A', 'Z')
```
