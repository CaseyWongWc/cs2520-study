# Mutable vs Immutable objects (from Objects lecture)

# IMMUTABLE (int, float, str, tuple) — 'changing' creates a NEW object
string_a = 'hello'
string_b = string_a       # both point to SAME object
string_b += ' world'      # creates a NEW string object
print(string_a)           # 'hello' — UNCHANGED!
print(string_b)           # 'hello world'

# MUTABLE (list, dict, set) — changes affect the original!
list_a = [1, 2, 3]
list_b = list_a           # both point to SAME list
list_b.append(4)          # modifies the ORIGINAL
print(list_a)             # [1, 2, 3, 4]  ← changed!
print(list_b)             # [1, 2, 3, 4]
