````python
#isinstance() with Custom Objects

#isinstance(obj, class) - a built-in python function to check if an object is an instance of a specified class

class Parent:
    pass

class Child(Parent):
    pass

class Other:
    pass

p = Parent()
c = Child()

print(isinstance(c, Child))
# ✅ True - c is an instance of Child class

print(isinstance(c, Parent))
# ✅ True - c is also an instance of Parent class (via inheritance!)

print(isinstance(p, Child))
# ❌ False - p is NOT an instance of Child class

print(isinstance(c, Other))
# ❌ False - c has no relationship at all
````

````python
#issubclass() - a built-in function to check if a class is a direct or indirect child of another class

class Parent:
    pass

class Child(Parent):
    pass

class GrandChild(Child):
    pass

class Other:
    pass

print(issubclass(Child, Parent))
# ✅ True - direct subclass

print(issubclass(GrandChild, Child))
# ✅ True - direct subclass

print(issubclass(GrandChild, Parent))
# ✅ True - indirect subclass!

print(issubclass(Parent, Child))
# ❌ False - Parent is a SUPER class, not a SUB class of Child

print(issubclass(Other, Parent))
# ❌ False - no relationship
````
